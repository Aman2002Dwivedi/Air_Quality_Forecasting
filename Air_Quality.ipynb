```python
import numpy as np
import pandas as pd

# --- Sub-index Functions Based on CPCB Guidelines ---
# Sample function for PM2.5 AQI
def get_pm25_aqi(pm):
    if pm <= 30:
        return np.interp(pm, [0, 30], [0, 50])
    elif pm <= 60:
        return np.interp(pm, [31, 60], [51, 100])
    elif pm <= 90:
        return np.interp(pm, [61, 90], [101, 200])
    elif pm <= 120:
        return np.interp(pm, [91, 120], [201, 300])
    elif pm <= 250:
        return np.interp(pm, [121, 250], [301, 400])
    elif pm <= 500:
        return np.interp(pm, [251, 500], [401, 500])
    else:
        return 500
    
def get_pm10_aqi(pm10):
    if pm10 <= 50:
        return np.interp(pm10, [0, 50], [0, 50])
    elif pm10 <= 100:
        return np.interp(pm10, [51, 100], [51, 100])
    elif pm10 <= 250:
        return np.interp(pm10, [101, 250], [101, 200])
    elif pm10 <= 350:
        return np.interp(pm10, [251, 350], [201, 300])
    elif pm10 <= 430:
        return np.interp(pm10, [351, 430], [301, 400])
    elif pm10 <= 500:
        return np.interp(pm10, [431, 500], [401, 500])
    else:
        return 500

def get_no2_aqi(no2):
    if no2 <= 40:
        return np.interp(no2, [0, 40], [0, 50])
    elif no2 <= 80:
        return np.interp(no2, [41, 80], [51, 100])
    elif no2 <= 180:
        return np.interp(no2, [81, 180], [101, 200])
    elif no2 <= 280:
        return np.interp(no2, [181, 280], [201, 300])
    elif no2 <= 400:
        return np.interp(no2, [281, 400], [301, 400])
    elif no2 <= 500:
        return np.interp(no2, [401, 500], [401, 500])
    else:
        return 500

def get_so2_aqi(so2):
    if so2 <= 40:
        return np.interp(so2, [0, 40], [0, 50])
    elif so2 <= 80:
        return np.interp(so2, [41, 80], [51, 100])
    elif so2 <= 380:
        return np.interp(so2, [81, 380], [101, 200])
    elif so2 <= 800:
        return np.interp(so2, [381, 800], [201, 300])
    elif so2 <= 1600:
        return np.interp(so2, [801, 1600], [301, 400])
    elif so2 <= 2000:
        return np.interp(so2, [1601, 2000], [401, 500])
    else:
        return 500

def get_co_aqi(co):
    if co <= 1:
        return np.interp(co, [0, 1], [0, 50])
    elif co <= 2:
        return np.interp(co, [1.1, 2], [51, 100])
    elif co <= 10:
        return np.interp(co, [2.1, 10], [101, 200])
    elif co <= 17:
        return np.interp(co, [10.1, 17], [201, 300])
    elif co <= 34:
        return np.interp(co, [17.1, 34], [301, 400])
    elif co <= 50:
        return np.interp(co, [34.1, 50], [401, 500])
    else:
        return 500

def get_ozone_aqi(o3):
    if o3 <= 50:
        return np.interp(o3, [0, 50], [0, 50])
    elif o3 <= 100:
        return np.interp(o3, [51, 100], [51, 100])
    elif o3 <= 168:
        return np.interp(o3, [101, 168], [101, 200])
    elif o3 <= 208:
        return np.interp(o3, [169, 208], [201, 300])
    elif o3 <= 748:
        return np.interp(o3, [209, 748], [301, 400])
    else:
        return 500

def get_nh3_aqi(nh3):
    if nh3 <= 200:
        return np.interp(nh3, [0, 200], [0, 100])  # WHO has no strict AQI for NH3, this is a rough mapping
    elif nh3 <= 400:
        return np.interp(nh3, [200, 400], [101, 200])
    elif nh3 <= 800:
        return np.interp(nh3, [401, 800], [201, 300])
    elif nh3 <= 1200:
        return np.interp(nh3, [801, 1200], [301, 400])
    elif nh3 <= 1800:
        return np.interp(nh3, [1201, 1800], [401, 500])
    else:
        return 500
```

```python
df['pm25_aqi'] = df['PM2.5 (ug/m3)'].apply(get_pm25_aqi)
df['pm10_aqi'] = df['PM10  (ug/m3)'].apply(get_pm10_aqi)
df['no2_aqi'] = df['NO2  (ug/m3)'].apply(get_no2_aqi)
df['so2_aqi'] = df['SO2  (ug/m3)'].apply(get_so2_aqi)
df['co_aqi'] = df['CO'].apply(get_co_aqi)
df['ozone_aqi'] = df['Ozone'].apply(get_ozone_aqi)
df['nh3_aqi'] = df['NH3  (ug/m3)'].apply(get_nh3_aqi)
```

```python
aqi_columns = ['pm25_aqi', 'pm10_aqi', 'no2_aqi', 'so2_aqi', 'co_aqi', 'ozone_aqi', 'nh3_aqi']
df['AQI'] = df[aqi_columns].max(axis=1).round(2)
```

```python
# Merge forecasted values with original AQI data
merged = forecast[['ds', 'yhat']].merge(df_aqi, on='ds', how='inner')  # keeps only overlapping dates

# Drop any missing values just in case
merged.dropna(inplace=True)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Actual and predicted values
y_true = merged['y']
y_pred = merged['yhat']

# Metrics
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2 = r2_score(y_true, y_pred)

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
import optuna
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics

def objective(trial):
    cp_scale = trial.suggest_float('changepoint_prior_scale', 0.001, 1.0, log=True)
    seas_scale = trial.suggest_float('seasonality_prior_scale', 1.0, 30.0)
    seasonality_mode = trial.suggest_categorical('seasonality_mode', ['additive', 'multiplicative'])
    yearly_fourier = trial.suggest_int('yearly_fourier_order', 5, 25)
    monthly_fourier = trial.suggest_int('monthly_fourier_order', 3, 15)

    model = Prophet(
        changepoint_prior_scale=cp_scale,
        seasonality_prior_scale=seas_scale,
        seasonality_mode=seasonality_mode,
        yearly_seasonality=False
    )

    # Add custom seasonalities
    model.add_seasonality(name='yearly', period=365.25, fourier_order=yearly_fourier)
    model.add_seasonality(name='monthly', period=30.5, fourier_order=monthly_fourier)

    # Fit model
    model.fit(df_aqi)

    # Cross-validation
    df_cv = cross_validation(model, initial='300 days', period='30 days', horizon='20 days', parallel="processes")
    df_p = performance_metrics(df_cv)
    
    return df_p['rmse'].mean()  # Objective: minimize RMSE
```

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=30)  # You can increase n_trials for better results
```

```python
print("Best parameters:")
print(study.best_params)

print("Best RMSE:", study.best_value)
```
